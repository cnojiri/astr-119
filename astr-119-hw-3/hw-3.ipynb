```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)    #create an array that is from 1 to 2*pi with 1000 values in bewtween

y = 5.5*np.cos(2*x)+5.5          #a--y(x) = 5.5 cos(2 * x) + 5.5

z = 0.02*(np.exp(x))             #b--y(x) = 0.02 * exp(x)

w = 0.25*(x**2)+0.1*np.sin(10*x) #c--y(x) = 0.25 * x^2 + 0.1 sin(10*x)
```

```python
#we are now plotting and labling the above equations
plt.plot(x,y, label=r'$y = 5.5 * cos(2 * x) +5.5$')        #make a plot of a, w the label of the equation
plt.plot(x,z, label=r'$y = 0.02 * exp(x)$')                #make a plot of b, w the label of the equation
plt.plot(x,w, label=r'$y = 0.25 * x^2 + 0.1 sin(10 * x)$') #make a plot of c, w the label of the equation

#adding labels for the x and y axis
plt.xlabel('Time in ASTR 119')       #labling the x-axis

plt.ylabel('Measures of Awesomenss') #labling the y-axis

#limits for the x and y values
plt.xlim([0,2*np.pi])                #this is the limit for x

plt.ylim([-1,10])                    #this is the limit for y

#adding legend
plt.legend(loc=1,framealpha=0.95)  #adding in the legend with a semi-transparet frame in upper RH corner
```
